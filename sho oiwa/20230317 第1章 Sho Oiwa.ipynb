```python
# classの例
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")
    
    
    def hello(self):
        print("Hello" + self.name + "!")
    
    
    def goodbye(self):
        print("Good-bye" + self.name + "!")


m = Man("David")
m.hello()
m.goodbye()
```

```python
# numpyの基本動作
import numpy as np

x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])

# 要素数が同じでないといけない
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x / 2.0)  # ブロードキャスト
```

```python
# numpy N次元配列
A = np.array([[1, 2],[3, 4]])
B = np.array([[3, 0],[0, 6]])
print(A.shape)
print(A)
print(B)
# 変わらず，同じ位置どうしの計算
print(A + B)
print(A * B)
print(A * 10)  # ブロードキャスト
```

```python
# numpy 要素へのアクセス
X = np.array([[51, 55], [14, 19], [0, 4]])
print(X)

# 参照もリストと変わらない
print(X[0])
print(X[0][1])

# Xを一次元配列に変換
X = X.flatten()
print(X)

#こんな参照もできる，但し，X[0, 2, 4]は通用しない
print(X[np.array([0, 2, 4])])
```

```python
# Matplotlib
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)  # numpyのrange関数
y = np.sin(x)

plt.plot(x, y)  # グラフ描画
plt.show()  # グラフ表示
```

```python
# pyplotの機能
y1 = np.sin(x)
y2 = np.cos(x)

# グラフの描画
plt.plot(x, y1, label = "sin")
plt.plot(x, y2, label = "cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend()  # legendは上のような凡例を表示するための処理
plt.show()
```
